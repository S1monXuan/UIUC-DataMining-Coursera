```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
url = 'categories.csv'
data = pd.read_csv(url)
data = data['Names'].str.split(';', expand=True)
data.head()
```

```python
files = []
size = len(data)
for i in range(0, size):
    row = data.iloc[i]
    cur = []
    for item in row:
        if item == None:
            break
        cur.append(item)
    files.append(cur)
files
```

```python
te = TransactionEncoder()
te_ary = te.fit(files).transform(files)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```
